```python
#Lets write a function name compare and pass two function if x is greater its call add function and 
#and operated addition else if it is smaller call substract and operate substarction.

def sum(x,y):
    return x+y
def substract(x,y):
    return x-y
def compare(x,y):
    if(x>y):
        print(sum(x,y))
    else:
        print(substract(x,y))
compare(10,2)

#Here x is greater therefore it call sum functon and operate additon.
```

```python
#Wap to find factorial using recursion.
def fact(x):
    if(x==0):
        return 1
    else:
        return x*fact(x-1)
fact(5)

#5*4*3*2*1
#20*6
```

```python
#Write a program which print sum of the factorial. (EX fact(5) = 5+4+3+2+1=15)
def fact(x):
    if(x==0):
        return 0
    else:
        return x+fact(x-1)
fact(5)
```

```python
#Same question another Method
def fact(x):
    if(x==0):
        return 0
    elif(x==1):
        return 1
    else:
        return x+fact(x-2)
fact(5)
    
```
